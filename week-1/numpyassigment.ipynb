#### numpy assignment

```python
import numpy as np
print(np.__version__)
np.show_config()

# null vector of size 10
Z = np.zeros(10)
print("Z =", Z)

# Ensure numpy is available in this cell
a = np.zeros(10, dtype=np.float64)
a.nbytes           # total bytes

#Null vector, 5th value = 1
b = np.zeros(10)
b[4] = 1
print(b)

#Vector 10→49
c = np.arange(10, 50)
print(c)

#Reverse vector
d = np.arange(50, 9, -1)
print(d)
print(d[::-1])  # or d = d[::-1]
print(d)

#3x3 matrix with values 0→8
e = np.arange(9).reshape(3, 3)
print(e)

#Identity matrix 3x3
f = np.eye(3)
print(f)

#3x3 matrix with random values
g = np.random.random((3, 3))
print(g)

#3x3x3 array with random values
h = np.random.random((3, 3, 3))
print(h)

#10×10 random; min & max
i = np.random.random((10, 10))
A_min, A_max = i.min(), i.max()
print(i)

#Random vector (30); mean
j = np.random.random(30)
m = j.mean()
print(j)
print(m)

#2D array with 1s on border, 0 inside
n = 6  # example
B = np.ones((n, n), dtype=int)
B[1:-1, 1:-1] = 0

#Pad array with a border of 0s
C = np.ones((3, 3))
C_pad = np.pad(C, pad_width=1, mode='constant', constant_values=0)

#Evaluate:
0 * np.nan            # -> nan
np.nan == np.nan      # -> False
np.inf > np.nan       # -> False
np.nan - np.nan       # -> nan
np.nan in set([np.nan])  # -> True  (same object identity in this expression)
0.3 == 3 * 0.1        # -> False (binary floating-point rounding)


#8×8 checkerboard
D = np.zeros((8, 8), dtype=int)
D[1::2, ::2] = 1
D[::2, 1::2] = 1
print(D)

#Index (x,y,z) of the 100th element in shape (6,7,8)
E = np.unravel_index(99, (6, 7, 8))
print(E)











```
