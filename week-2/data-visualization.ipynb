#### Line Chart → Trends over time

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data for the birds DataFrame
data = {
	'species': ['Snow Geese', 'Snow Geese', 'Snow Geese', 'Other Bird', 'Other Bird'],
	'year': [2000, 2001, 2002, 2000, 2001],
	'count': [100, 150, 200, 50, 60]
}
birds = pd.DataFrame(data)

snow_geese = birds[birds['species'] == 'Snow Geese']
snow_geese_grouped = snow_geese.groupby('year')['count'].sum()

plt.figure(figsize=(8,5))
plt.plot(snow_geese_grouped.index, snow_geese_grouped.values, marker='o')
plt.title("Snow Geese Population Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

#### Scatter Plot → Relationship between two factors

```python
plt.figure(figsize=(8,5))
plt.scatter(snow_geese['year'], snow_geese['count'], alpha=0.6)
plt.title("Snow Geese Sightings Scatter")
plt.xlabel("Year")
plt.ylabel("Count per Observation")
plt.show()
```

#### Bar Chart → Compare across categories

```python
# Ensure the cell defining 'birds' DataFrame is executed before running this cell

# Adding a dummy 'region' column to the birds DataFrame for demonstration
import numpy as np  # Ensure numpy is imported
birds['region'] = np.random.choice(['North', 'South', 'East', 'West'], size=len(birds))

# Grouping by 'region' and summing the counts
top_regions = birds.groupby('region')['count'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
top_regions.plot(kind='bar', color='skyblue')
plt.title("Top 5 Regions with Snow Geese Sightings")
plt.xlabel("Region")
plt.ylabel("Total Count")
plt.xticks(rotation=45)
plt.show()
```
